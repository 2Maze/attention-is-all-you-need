```python
import torch
import matplotlib.pyplot as plt

from model.embeddings import TransformerEmbedding
```

```python
EMB_DIM = 256
MAX_SEQ_SIZE = 60

emb = TransformerEmbedding(vocab_size=100, dim_model=EMB_DIM, max_len=MAX_SEQ_SIZE)
```
